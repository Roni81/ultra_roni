```python
class Caculator:
    def __init__(self): #__init__  class -> 객체 생성될 때 자동 호출(생성자)
        self.result = 0
        print("생성자 호출됨")
        
    def add(self, num):
        self.result += num
        print("add함수 호출됨")
        return self.result
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
x_train.shape
```

```python
W = torch.zeros(1, requires_grad = True) #requires_grad = True는 W는 변수
```

```python
b = torch.zeros(1, requires_grad = True) #requires_grad = True는 b는 변수
```

```python
#미분을 해서 구한 기울기를 0으로 초기화 해줌
# optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
# cost.backward() #W, b에 대한 기울기가 계산
# optimizer.step()#W,b에 대한 업데이트
```

```python
for epoch in range(2000):
    hx=W * x_train + b
    cost=torch.mean((hx - y_train)**2)
    optimizer=optim.SGD([W,b],lr=0.01) #lr == learning_rate
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%100 ==0:
        print(epoch,  W.item(), b.item(), cost.item())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
model = nn.Linear(1,1) #입력차원, 출력차원
```

```python
list(model.parameters()) #w, b 값
#초기모델 : y = 0.5153 * x_train -0.4414
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
for epoch in range(2000):
    hx = model(x_train) #hx=x_train.matmul(W) + b
    cost = F.mse_loss(hx, y_train) #cost=torch.mean ((hx - y_train)**2)

    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%100 ==0:
        print(epoch, cost.item())
```
