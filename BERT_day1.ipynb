```python
from transformers import TFBertForNextSentencePrediction
```

```python
from transformers import AutoTokenizer
```

```python
model = TFBertForNextSentencePrediction.from_pretrained('bert-base-uncased')
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
prompt = "In Italy, pizza served in formal settings, such as at a restaurant, is presented unsliced."
next_sentence = "The sky is blue due to the shorter wavelength of blue light."
```

```python
encoding = tokenizer(prompt, next_sentence, return_tensors='tf')
```

```python
encoding
```

```python
pip( "In Italy, [MASK] served in formal settings, such as at a restaurant, is presented unsliced.")
```

```python
pip('in Korea, [MASK] served in formal setting, such as at a home')
```

```python
model(encoding['input_ids'], encoding['token_type_ids'])
```

```python
logits = model(encoding['input_ids'], encoding['token_type_ids'])[0]
```

```python
import tensorflow as  tf
```

```python
print(tf.keras.layers.Softmax(logits))
```

```python
soft = tf.keras.layers.Softmax()
res = soft(logits)
```

```python
res
```

```python
tf.math.argmax(res, axis = -1).numpy()
#0이면 이어지는 문장
```
